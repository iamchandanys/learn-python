Indentation

```python
# Define a function with proper indentation
def greet(name):
    if name:
        print(f"Hello, {name}!")
    else:
        print("Hello, World!")

# Call the function with the variable 'name'
greet('Welcome!')
```

Line Continuation

```python
# Example of line continuation using backslash
long_string = "This is a very long string that we want to split " \
              "into multiple lines for better readability."

print(long_string)
```

Multiple Statements on a single line

```python
x = 10; y = 20; z = x + y; print(f"x: {x}, y: {y}, z: {z}")
```

Variable Assignments and Types

```python
# Single assignment
a = 5

# Multiple assignments
b, c, d = 1, 2, 3

# Assigning the same value to multiple variables
e = f = g = 10

# Unpacking a list
h, i, j = [4, 5, 6]

# Unpacking a tuple
k, l, m = (7, 8, 9)

# Printing the variables to verify the assignments
print(f"a: {a}")
print(f"b: {b}, c: {c}, d: {d}")
print(f"e: {e}, f: {f}, g: {g}")
print(f"h: {h}, i: {i}, j: {j}")
print(f"k: {k}, l: {l}, m: {m}")

print(type(a))
```

```python
# We can also changes the type of a variable, for example:

a = 5
print(type(a))  # <class 'int'>
a = 'Hello, World!'
print(type(a))  # <class 'str'>
```
