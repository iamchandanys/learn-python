```python
# Assignment 1 to 4

user_input = float(input("Enter the number: "))

if user_input > 0:
    print(f"The number {user_input} is positive")
elif user_input < 0:
    print(f"The number {user_input} is negative")
else:
    print(f"The number is zero")

# If the number is positive
if user_input > 0:
    if user_input % 2 == 0:
        print(f"The number {user_input} is positive and even")
    else:
        print(f"The number {user_input} is positive but odd")
elif user_input == 0:
    print(f"The number is zero.")
```

```python
# Assignnment 7

for row in range(5):
    for col in range(5):
        # The end=" " argument in the print() function controls how the output is terminated after printing each item.
        # Normally, print() adds a new line (\n) at the end of every print statement
        # When we use end=" ", it replaces the default newline (\n) with a space (" "), so the output stays on the same line with a space after each *
        print("*", end=" ")
    # print() moves to the next line after completing a row
    print()
```

```python
# Assignment 8

count = 0

while True:
    usr_input = int(input("Enter the number: "))
    if(usr_input == 0):
        # Break will terminate the loop
        break
    else:
        count += user_input

print(count)
```

```python
# Assignment 12
# Example: 5! -> 5*4*3*2*1 = 120

users_input = input("Enter the number")
index = 1
count = 1

while index <= int(users_input):
    count = count * index
    index += 1

print(count)
```

```python
# Assignment 13
# Example: 5 -> 1+2+3+4+5 = 15

users_input = input("Enter the number")
index = 0
count = 0

while index <= int(users_input):
    count = count + index
    index += 1

print(count)
```
