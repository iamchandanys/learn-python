**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# What is a Factorial? The factorial of a number is the multiplication of all positive integers less than or equal to that number. It is denoted by the symbol !.
# Example: The factorial of 5 (written as 5! or n!) is calculated as: 5! = 5 × 4 × 3 × 2 × 1 = 120

# Using Recursion

def calculate_factorial(user_input):
    if user_input == 0 or user_input == 1:
        return 1
    else:
        return user_input * calculate_factorial(user_input - 1)
    
print(calculate_factorial(5)) # 120

# Using a For Loop and Range

def calculate_factorial_using_loop(n):
    result = 1
    for i in range(2, n + 1): # Loop from 2 to n
        result *= i  # Multiply result by i in each iteration
    return result

print(calculate_factorial_using_loop(5)) # 120
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
# Formula: n * (n + 1) / 2
# For example, the sum of the first 5 natural numbers (1 + 2 + 3 + 4 + 5) is: 5 * (5 + 1) / 2 = 15

def calculate_first_n_natural_number(num):
    return num * (num + 1) // 2

calculate_first_n_natural_number(5)
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
def is_leap_year(year):
    if not isinstance(year, int):  # Check if year is an integer
        raise TypeError("Year must be an integer.")

    if year < 0:  # Handle negative years (not typically considered in leap year rules)
        raise ValueError("Year must be a non-negative integer.")

    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
    
is_leap_year(2028)
```
