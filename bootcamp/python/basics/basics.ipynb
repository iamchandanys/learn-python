Indentation

```python
# Define a function with proper indentation
def greet(name):
    if name:
        print(f"Hello, {name}!")
    else:
        print("Hello, World!")

# Call the function with the variable 'name'
greet('Welcome!')
```

Variable Assignments and Types

```python
# Single assignment
a = 5

# Multiple assignments
b, c, d = 1, 2, 3

# Assigning the same value to multiple variables
e = f = g = 10

# Unpacking a list
h, i, j = [4, 5, 6]

# Unpacking a tuple
k, l, m = (7, 8, 9)

# Printing the variables to verify the assignments
print(f"a: {a}")
print(f"b: {b}, c: {c}, d: {d}")
print(f"e: {e}, f: {f}, g: {g}")
print(f"h: {h}, i: {i}, j: {j}")
print(f"k: {k}, l: {l}, m: {m}")

print(type(a))
```

```python
# We can also changes the type of a variable, for example:

a = 5
print(type(a))  # <class 'int'>
a = 'Hello, World!'
print(type(a))  # <class 'str'>

# Note: The type of the variable is determined during runtime, not during compile time.
```

Variable Naming Conventions

```python
# Variable naming conventions

# Use descriptive names
user_age = 25
user_name = "Alice"

# Use underscores to separate words
total_price = 100.50

# Avoid using reserved keywords
# for = 5  # This will cause a syntax error

# Use lowercase letters for variable names
item_count = 10

# Use uppercase letters for constants
PI = 3.14159

# Avoid using single character names except for loop variables
x = 5  # Not recommended
for i in range(10):  # Acceptable for loop variables
    print(i)

# Follow the same conventions for function names
def calculate_total(price, tax_rate):
    return price + (price * tax_rate)

total = calculate_total(total_price, 0.08)
print(total)
```

Type Conversion

```python
# Integer to float
int_value = 10
float_value = float(int_value)
print(f"Integer to float: {float_value} (type: {type(float_value)})")

# Float to integer
float_value = 10.75
int_value = int(float_value)
print(f"Float to integer: {int_value} (type: {type(int_value)})")

# String to integer
str_value = "123"
int_value = int(str_value)
print(f"String to integer: {int_value} (type: {type(int_value)})")

# String to float
str_value = "123.45"
float_value = float(str_value)
print(f"String to float: {float_value} (type: {type(float_value)})")

# Integer to string
int_value = 123
str_value = str(int_value)
print(f"Integer to string: {str_value} (type: {type(str_value)})")

# Float to string
float_value = 123.45
str_value = str(float_value)
print(f"Float to string: {str_value} (type: {type(str_value)})")

# List to tuple
list_value = [1, 2, 3]
tuple_value = tuple(list_value)
print(f"List to tuple: {tuple_value} (type: {type(tuple_value)})")

# Tuple to list
tuple_value = (1, 2, 3)
list_value = list(tuple_value)
print(f"Tuple to list: {list_value} (type: {type(list_value)})")
```

Inputs

```python
# Taking input from the user

# Input as a string
name = input("Enter your name: ")
print(f"Hello, {name}!")

# Input as an integer
age = int(input("Enter your age: "))
print(f"You are {age} years old.")

# Input as a float
height = float(input("Enter your height in meters: "))
print(f"Your height is {height} meters.")

# Input as a list of integers
numbers = list(map(int, input("Enter a list of numbers separated by space: ").split()))
print(f"The list of numbers is: {numbers}")
```

Math Operators

```python
"""
This script shows basic math operations in Python, including:
- Addition: Adds two numbers.
- Subtraction: Subtracts one number from another.
- Multiplication: Multiplies two numbers.
- Division: Divides one number by another.
- Modulus: Finds the remainder of division.
- Exponentiation: Raises a number to a power.
- Floor Division: Divides and rounds down to the nearest whole number.
"""

# Addition
sum_result = 10 + 6
print(f"10 + 6 = {sum_result}")

# Subtraction
sub_result = 10 - 5
print(f"10 - 5 = {sub_result}")

# Multiplication
mul_result = 10 * 5
print(f"10 * 5 = {mul_result}")

# Division
div_result = 10 / 3
print(f"10 / 3 = {div_result}")

# Modulus
mod_result = 10 % 3
print(f"10 % 3 = {mod_result}")

# Exponentiation
exp_result = 10 ** 2
print(f"10 ** 2 = {exp_result}")

# Floor Division
floor_div_result = 10 // 3
print(f"10 // 3 = {floor_div_result}")
```

Comparision Operators

```python
# Variables for demonstration
x = 10
y = 20

# Equal to
equal_result = (x == y)
print(f"x == y: {equal_result}")

# Not equal to
not_equal_result = (x != y)
print(f"x != y: {not_equal_result}")

# Greater than
greater_than_result = (x > y)
print(f"x > y: {greater_than_result}")

# Less than
less_than_result = (x < y)
print(f"x < y: {less_than_result}")

# Greater than or equal to
greater_than_equal_result = (x >= y)
print(f"x >= y: {greater_than_equal_result}")

# Less than or equal to
less_than_equal_result = (x <= y)
print(f"x <= y: {less_than_equal_result}")
```

Logical Operators

```python
# Logical Operators in Python

# Variables for demonstration
a = True
b = False

# Logical AND
and_result = a and b
print(f"a and b: {and_result}")

# Logical OR
or_result = a or b
print(f"a or b: {or_result}")

# Logical NOT
not_result = not a
print(f"not a: {not_result}")

# Combining logical operators
combined_result = (a and b) or (not b)
print(f"(a and b) or (not b): {combined_result}")
```
