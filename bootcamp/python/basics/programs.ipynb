Get Factorial of a number using For Loop

```Example: 5! -> 5*4*3*2*1 = 120```

````python
def factorial_using_for_loop(n):
    count = 1
    for num in range(n, 0, -1):
        count *= num
    return count

factorial_using_for_loop(5)
````

Get Factorial of a number using While Loop

```Example: 5! -> 5*4*3*2*1 = 120```

````python
def factorial_using_while_loop(n):
    index = n
    count = 1
    while index >= 1:
        count *= index
        index = index - 1
    return count

factorial_using_while_loop(5)
````

Print all Prime Numbers from 1 to 20

````python
def print_all_prime_numbers(n):
    for num in range(1, n + 1):
        if(num % 2 == 0):
            print(num, end=" ")

print_all_prime_numbers(20)
````

Check whether the given year is Leap Year?

````python
def is_leap_year(n):
    if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0):
        return True
    else:
        return False

is_leap_year(2025)
````
