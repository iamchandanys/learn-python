```python
# Study string basics: single quotes vs double quotes vs triple quotes — when do you use each?

# Single quotes — basic string
name = 'bangalore'

# Double quotes — use when string has an apostrophe
msg = "it's a beautiful day"

# Triple quotes — for multi-line strings
description = """
This is
a multi-line string.
"""

print(name)
print(msg)
print(description)
```

```python
# Study and practice 8 string methods: 
# .upper(), .lower(), .strip(), .split(), .replace(), .startswith(), .endswith(), .find()
# call each on a sample string and print the result

#All uppercase
print("baNgalore".upper())

# All lowercase
print("baNgalore".lower())

# Removes surrounding whitespace
print(" chandan kumar   ".strip())

# Splits into a list (by whitespace or separator)
print("one,two,three".split(","))

# Swaps a substring for another
print("Hello World".replace("World", "Python"))

# Checks the beginning — returns True/False
print("Hello World".startswith("Hello"))

# Checks the end — returns True/False
print("Hello World".endswith("World"))

# Returns the index of the first occurrence of a substring. 
# Returns -1 if not found. Indexing starts at 0.
print("Hello World".find("World"))
```

```python
# Study f-strings: 
# Write a function greet(name, age) that returns "Hello, {name}! You are {age} years old." using an f-string.

def greet(name, age):
    return f"Hello, {name}! You are {age} years old."

print(greet("Chandan", 30))
```
