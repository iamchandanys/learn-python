#### Study string basics: single quotes vs double quotes vs triple quotes — when do you use each?

```python
# Single quotes — basic string
name = 'bangalore'

# Double quotes — use when string has an apostrophe
msg = "it's a beautiful day"

# Triple quotes — for multi-line strings
description = """
This is
a multi-line string.
"""

print(name)
print(msg)
print(description)
```
