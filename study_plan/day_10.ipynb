```python
# Declare variables of each type: int, float, str, bool, None

_int = 10
print(_int, type(_int))

_float = 10.30
print(_float, type(_int))

_str = "bangalore"
print(_str, type(_int))

_bool = True
print(_bool, type(_int))

_none = None
print(_none, type(_none))
```

```python
# Study type conversion. Run each and note which ones raise ValueError and why

print(str(_int))
# print(int(_str)) <- Raises Value Error
print(int(_bool))
print(str(_none))
```

```python
# Check variable type using isinstance() (preferred) or type()

if isinstance(_int, int):
    print("it's an int")

if type(_int) == int:
    print("it's an int")
```

```python
# Write a function describe(value) that prints whether the input is an int, a string, a float, or “unknown type” — test with 5 different inputs

def describe(value):
    if isinstance(value, int):
        label = "int"
    elif isinstance(value, float):
        label = "float"
    elif isinstance(value, str):
        label = "string"
    elif isinstance(value, bool):
        label = "bool"
    else:
        label = "unknown"

    print(f"The given input is {label}")

describe(10)
describe(10.30)
describe("bangalore")
describe(True)
describe(None)
```
