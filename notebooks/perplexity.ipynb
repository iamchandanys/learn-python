To get the index of for loop

```python
cities = ["Berlin", "Paris", "London", "Madrid", "Rome"]

for index, city in enumerate(cities):
    print(f"{index}: {city}")
```

Swap two variables without using a third variable

```python
a = 10
b = 20

a, b = b, a

print(a, b) # Output: 20 10
```

List comprehension - Filtering Even Numbers - Any number that can be divided by 2 with no remainder

```python
[x for x in range(10) if x % 2 == 0]
```

List comprehension - Multiples of 3 - Any number that can be divided by 3 with no remainder

```python
[x for x in range(1, 21) if x % 3 == 0]
```

Find duplicate elements in a list using a set

```python
nums = [1, 2, 3, 2, 4, 5, 1, 6]

set([x for x in nums if nums.count(x) > 1])
```

Write a Python function that, given a list of numbers and a target value, finds **two different numbers in the list that add up to the target**. If such a pair exists, return the indices (positions) of those two numbers as a tuple (for example, (0, 1)). If no pair is found, return None.

```python
nums = [2, 7, 11, 15]
target = 9
# Expected output: (0, 1)

print(len(nums))

def findPositions(arrOfNums: list[int], targetVal: int):
    print([x for idx, x in enumerate(arrOfNums) if x + (nums[idx+1])] == targetVal)

findPositions(nums, target)
```
