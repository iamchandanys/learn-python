### Find the duplicate numbers from the list

```python
x = [22, 56, 35, 22, 88, 56, 95, 65]
dup_nums = []

for num in x:
    dup_nums.append(num) if x.count(num) > 1 else None
    
print(list(set(dup_nums)))
```

### Frequency Counter

Check if two numbers have the same digits appearing the same number of times.

**Simple explanation:** 
- Take two numbers and count how many times each digit appears in both numbers
- If both numbers have exactly the same digits with the same counts, return True
- Otherwise, return False

**Examples:**
- 224 and 242: Both have two 2's and one 4, so they match ✓
- 34 and 14: First has 3,4 and second has 1,4 - different digits ✗
- 22 and 222: First has two 2's, second has three 2's - different counts ✗

**Requirements:** 
- Time complexity must be O(N) where N is the number of digits

```python
def same_frequency(num1, num2) -> bool:
    # Convert numbers to strings to work with digits
    num1_str = str(num1)
    num2_str = str(num2)
    
    # If lengths are different, they can't have same frequency
    if len(num1_str) != len(num2_str):
        return False
    
    # Count frequency of each digit in both numbers
    freq1 = {}
    freq2 = {}
    
    # Count digits in first number
    for digit in num1_str:
        freq1[digit] = freq1.get(digit, 0) + 1
    
    # Count digits in second number
    for digit in num2_str:
        freq2[digit] = freq2.get(digit, 0) + 1
    
    # Compare the frequency dictionaries
    return freq1 == freq2

# Test cases
print(same_frequency(224, 242)) # True - both have 2 twos and 1 four
print(same_frequency(34, 14)) # False - different digits (3 vs 1)
print(same_frequency(3589578, 5879385)) # True - same digits, same counts
print(same_frequency(22, 222)) # False - different counts of 2
```

### Missing in Array

You have a list of numbers that should contain all numbers from 1 to n, but one number is missing. Find which number is missing.

```python
def missing_num(arr: list):
    set_arr = set(arr)
    
    for x in range(min(arr), max(arr) + 1):
        if x not in set_arr:
            return x
        
    return max(arr) + 1

print(missing_num([1, 2, 3, 5])) # 4
print(missing_num([8, 2, 4, 5, 3, 7, 1])) # 6
print(missing_num([1])) # 2
```
